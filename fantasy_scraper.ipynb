```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# URL of the page
url = 'https://www.futbolfantasy.com/analytics/laliga-fantasy/puntos'

# Send a GET request
response = requests.get(url)
response.raise_for_status()  # Check for a successful request

# Parse the HTML
soup = BeautifulSoup(response.text, 'html.parser')

# Find all player rows starting with the class 'elemento elemento_jugador'
players = soup.find_all('div', class_=lambda value: value and value.startswith('elemento elemento_jugador'))

data = []
for player in players:
    
    player_name = player.find('span', class_='d-none d-lg-block').get_text(strip=True)
    team_name = player.find('span', class_='mb-auto').get_text(strip=True)
    points = player.find('span', class_='m-auto columna_puntos columna_puntostemporada').get_text(strip=True)

    last3 = player.find_all('span', class_= lambda value: value and value.startswith('racha-box'))
    try:
        match_tmin1 = last3[0].get_text(strip=True)
    except IndexError:
        match_tmin1 = None
    try:
        match_tmin2 = last3[1].get_text(strip=True)
    except IndexError:
        match_tmin2 = None
    try:
        match_tmin3 = last3[2].get_text(strip=True)
    except IndexError:
        match_tmin3 = None

    if match_tmin1 == '-':
        match_tmin1 = None
    
    if match_tmin2 == '-':
        match_tmin2 = None
    
    if match_tmin3 == '-':
        match_tmin3 = None

    data.append({
        'Player Name': player_name,
        'Team Name': team_name,
        'Total Points': int(points),
        'Points_T-1':match_tmin1,
        'Points_T-2':match_tmin2,
        'Points_T-3':match_tmin3
    })
```

```python
df = pd.DataFrame(data)
```
